```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_playstore.csv")
```

```python
from autoviz import AutoViz_Class
AV = AutoViz_Class()

filename = "cleaned_playstore.csv"
target_variable = "autoviz"
custom_plot_dir = "autoviz"

dft = AV.AutoViz(
    filename,
    sep=",",
    depVar=target_variable,
    dfte=None,
    header=0,
    verbose=2,
    lowess=False,
    chart_format="html",
    max_rows_analyzed=150000,
    max_cols_analyzed=30,
    save_plot_dir=custom_plot_dir
)
```
